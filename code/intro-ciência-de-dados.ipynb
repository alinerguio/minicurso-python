```python
dados = dados.drop(columns=['data'])
dados = dados.drop(columns=['permissao'])
```

```python
dados.loc[dados['sabe_programar'] == 'Sim', 'sabe_programar'] = 1
```

```python
dados
```

```python
colunas_alterar = ['sabe_programar', 'conhece_python', 'usou_python', 'python_instalado']

for coluna in colunas_alterar:
    print(coluna)
    dados.loc[dados[coluna] == 'Sim', coluna] = 1
    dados.loc[dados[coluna] == 'Não', coluna] = 0
```
