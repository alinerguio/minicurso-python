## Explorar os dados visualmente
Gráficos 

```python
pip install matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
sabe_programar = dados['sabe_programar'].to_list()
nivel_programar = dados['nivel_programar'].to_list()
conhece_python = dados['conhece_python'].to_list()
usou_python = dados['usou_python'].to_list()
python_instalado = dados['python_instalado'].to_list()
```

```python
plt.hist(sabe_programar, color = 'skyblue')
```

```python
plt.hist(nivel_programar, color='pink')
```

```python
plt.plot(conhece_python, usou_python, 'o', color='red')
plt.plot(usou_python, python_instalado, '*')
plt.title('Relação de conhecimento de python')
```

```python
dados['pessoa'] = 1
```

```python
dados_sabe_programar = dados.groupby('sabe_programar').count().reset_index()
dados_sabe_programar
```

```python
dados_sabe_programar = dados_sabe_programar.drop(columns=['nivel_programar', 'so']) # dados não estão dizendo nada
dados_sabe_programar
```

```python
dados_sabe_programar.plot.bar(x='sabe_programar', y='pessoa', color='#030303')
```
